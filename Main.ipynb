```python


print('start init data');

heart_data =  EDA.get_clean_data()
X = heart_data.drop(columns='activity')
y = heart_data['activity']
activity = y.unique()
X_train_Imbalance, X_test_Imbalance, y_train_Imbalance, y_test_Imbalance = train_test_split(X, y, test_size=0.2, random_state=0)


scaler = MinMaxScaler()
X_train_scaler_Imbalance = scaler.fit_transform(X_train_Imbalance)
X_test_scaler_Imbalance = scaler.transform(X_test_Imbalance)
print('end scaler Imbalance data ');
print('start balance data with SMOTE');
smote = SMOTE()
X_train_balanced, y_train_balanced = smote.fit_resample(X, y)
X_train, X_test, y_train, y_test = train_test_split(X_train_balanced, y_train_balanced, test_size=0.2, random_state=0)
print('SMOTE ok');
scaler = MinMaxScaler()
X_train_scaler = scaler.fit_transform(X_train)
print('fit transform ok');
X_test_scaler = scaler.transform(X_test)

print('End')
```

vvvvvvvvvvvvv

```python

print("Start SVM Imbalance")
svm_Imbalance = SVC(C=3)
svm_Imbalance.fit(X_train_scaler_Imbalance, y_train_Imbalance)
print('svm fit ok');
# Predict on the testing set
svc_pred_Imbalance = svm_Imbalance.predict(X_test_scaler_Imbalance)
print('svm predict ok');
print(classification_report(y_test_Imbalance, svc_pred_Imbalance))

svc_conf_mx_Imbalance = confusion_matrix(y_test_Imbalance,svc_pred_Imbalance, labels = activity)
#print(svc_conf_mx)
f,ax = plt.subplots(figsize=(13,13))
sns.heatmap(svc_conf_mx_Imbalance, cmap="Greens", annot = True, cbar_kws={"orientation":"vertical","label":"color bar"})
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.savefig('svm_imbalance.png', dpi=300, bbox_inches='tight')
plt.show()
print("End SVM Imbalance")
```

vvvvvvvvvvvvvvvvvvvvvvvv

```python
print("Start SVM balance")
svc = SVC(C=3)

svc_model = svc.fit(X_train_scaler, y_train)
print('svm fit ok');
svc_pred = svc_model.predict(X_test_scaler)
print('svm predict ok');
print(classification_report(y_test, svc_pred, labels = activity))


svc_conf_mx = confusion_matrix(y_test,svc_pred, labels = activity)
#print(svc_conf_mx)
f,ax = plt.subplots(figsize=(13,13))
sns.heatmap(svc_conf_mx, cmap="Greens", annot = True, cbar_kws={"orientation":"vertical","label":"color bar"})
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.savefig('svm_balance.png', dpi=300, bbox_inches='tight')
plt.show()
print("End SVM balance")
```
